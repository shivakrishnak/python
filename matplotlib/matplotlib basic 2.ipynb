## Figure Size and DPI

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig,axes = plt.subplots(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# To save a figure to a file
fig.savefig('my_picture.jpg', dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Square')
ax.plot(x,x**3, label='X Cube')

ax.legend()
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#FF8C00', linewidth = 3, alpha = 0.5,linestyle='-')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#FF8C00', linewidth = 3, alpha = 0.5,linestyle='-', marker='o',
        markersize='15',markerfacecolor='red',markeredgewidth=3,markeredgecolor='black')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#FF8C00', lw = 3, alpha = 0.5,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
